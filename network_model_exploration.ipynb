```python
import keras
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
########################
##    Reading data    ##
########################

train_input = np.load('train_input.npy')
train_target = np.load('train_target.npy')

test_input = np.load('test_input.npy')
test_target = np.load('test_target.npy')

print('train_input  :', train_input.shape)
print('train_target :', train_target.shape, end='\n\n')
print('test_input   :', test_input.shape)
print('test_target  :', test_target.shape)
```

```python
#########################
##    Reading model    ##
#########################

model = keras.models.load_model('model_1.h5')
model.summary()
```

```python
#####################################################
##   Evaluating model on train and test datasets   ##
#####################################################

train_loss = model.evaluate(train_input, train_target)
test_loss = model.evaluate(test_input, test_target)

print('\n\nModel loss on train data    :', train_loss)
print('Model loss on test data     :', test_loss, end='\n\n\n')
```

```python
######################
##    Predicting    ##
######################

pred_train = model.predict(train_input)
pred_test = model.predict(test_input)
```

```python
##############################
##    Some train example    ##
##############################

plt.figure(figsize=(20, 5))
plt.plot(train_target[:,3,23])
plt.plot(pred_train[:, 3, 23])
plt.title('Train data')
plt.xlabel('Time')
plt.ylabel('Count_of_events')
plt.show()
```

```python
########################
##    Some example    ##
########################

plt.figure(figsize=(20,5))
plt.plot(test_target[:,3,23])
plt.plot(pred_test[:,3,23])
plt.title('Test data')
plt.xlabel('Time')
plt.ylabel('Count_of_events')
plt.show()
```
