```python
import os
import datetime
import numpy as np
import pandas as pd
```

# Here the final processing version is presented

```python
########################
##    Reading data    ##
########################

data = pd.read_pickle('selected_data.pkl')
print('data :', data.shape)
data.head()
```

```python
#################################################
##    One-Hot vectors for states and events    ##
#################################################

event_types = pd.get_dummies(data.EVENT_TYPE)
states = pd.get_dummies(data.STATE)
cost = data.DAMAGE_PROPERTY + data.DAMAGE_CROPS

#########################
##    Date features    ##
#########################

month = data.DATE.apply(lambda x: x.month)
day = data.DATE.apply(lambda x: x.day)
year = data.DATE.apply(lambda x: x.year)
```

```python
###############################
##    Merging the columns    ##
###############################

cost_data = event_types
cost_data[states.columns] = states

cost_data['COST'] = cost

cost_data['INJURIES_DIRECT'] = data.INJURIES_DIRECT
cost_data['INJURIES_INDIRECT'] = data.INJURIES_INDIRECT
cost_data['DEATHS_DIRECT'] = data.DEATHS_DIRECT
cost_data['DEATHS_INDIRECT'] = data.DEATHS_INDIRECT

cost_data['MONTH'] = month
cost_data['DAY'] = day
cost_data['YEAR'] = year
```

```python
#######################
##    Saving data    ##
#######################

cost_data.to_pickle('cost_data.pkl')
```
